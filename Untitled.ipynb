```python
#TASK 1.2

#Task 1.2: Investigate the data
#Use matplotlib.pyplot (meaning plt) to draw a barplot for the distribution of targets in y. Use the target_names from the dataset as labels for the x-axis.
#For this, define the funtion distribution_plot(data, target_names, figure_name) because we'll need it later on. This function must take three arguments: data, 
#the data to plot, target_names the names of the classes (a list-like object), and figure_name, a string that should appear on top of the whole figure.
#For this task, your function must be called with distribution_plot(y, IRIS.target_names, 'Total Data Distribution').
```

```python
def distribution_plot(data, target_names, figure_name):
    seti = 0
    versi = 0
    virgi = 0
    listi = []
    
    for value in data:
        if value == 0: 
            seti += 1
        elif value == 1:
            versi +=1
        elif value == 2:
            virgi +=1

    listi.append(seti)
    listi.append(versi)
    listi.append(virgi)

    plt.bar(target_names, listi)
    plt.title(figure_name)
    plt.show()
```

```python
distribution_plot(y, IRIS.target_names, 'Total Data Distribution')
```

```python
#TASK 1.3

#For a proper machine learning experiment, we must split the data (X and y from Task 1.1) into a mutually exclusive train and test set.

#Use the function train_test_split. For this task, you have to use the random_state argument and set it to 23. Otherwise, the test will fail. We want to use 70% of the data for training.

#Call the splits accordingly: X_train, y_train, X_test, y_test.

#Afterwards, call your distribution_plot function twice, namely with y_train and y_test for the train and test distribution.
```

```python
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=23)

distribution_plot(y_train, IRIS.target_names, 'Total Data Distribution')
distribution_plot(y_test, IRIS.target_names, 'Total Data Distribution')
```
